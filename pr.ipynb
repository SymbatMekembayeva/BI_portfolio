```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set_style("darkgrid")
plt.rc('figure', autolayout=True)
plt.rc('axes', labelweight='bold', labelsize='large', titleweight='bold', titlesize=18, titlepad=10)
```

```python
data=pd.read_csv("assignment2new.csv")
data.head()
```

```python
colors = ["#FF5733", "#33FF57", "#5733FF"]
sns.boxplot(x="demographics", y="time_spent", data=data, palette=colors)
plt.show()
```

```python
sns.histplot(data=data, x="time_spent", hue="gender", bins=15, kde=True)
plt.title('Histogram with Hue')
plt.show()
```

```python
colors = ["#FF5733", "#33FF57", "#5733FF", "#FFFF33", "#B19CD9"]
sns.barplot(x=data['interests'], y=data['time_spent'], palette=colors)
plt.show()
```

```python
sns.kdeplot(data=data, x="time_spent", hue="platform", shade=True)
plt.show()
```

```python
sns.swarmplot(x=data['location'], y=data['income'])
```

```python
plt.hist(data['income'])
plt.xlabel("income")
plt.show()
```
