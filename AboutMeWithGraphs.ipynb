```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Create a DataFrame with the provided data
data = {
    'Month': ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'June', 'July', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'],
    'Total': [74288, 109743, 110037, 128468, 136771, 144156, 223515, 280851, 137475, 186373, 73580, 70928],
    'Avg': [2396, 3919, 3549, 4282, 4411, 4805, 7210, 9059, 4582, 6012, 2452, 2288]
}

df = pd.DataFrame(data)

# Set the style of the plot
sns.set(style="whitegrid")

# Initialize the matplotlib figure
plt.figure(figsize=(12, 6))

# Plot total steps
sns.barplot(x="Month", y="Total", data=df, color="skyblue", label="Total amount")


# Add labels and title
plt.xlabel("Month", fontsize=14)
plt.ylabel("Amount of Steps", fontsize=14)
plt.title("Yearly Report of Steps made in each month", fontsize=16)

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

# Add legend
plt.legend(loc="upper left")

# Show plot
plt.tight_layout()
plt.show()
```

```python
months = ['October', 'January', 'March', 'May', 'June']
test_scores = [69, 101, 116, 128, 129]

score_changes = [test_scores[i+1] - test_scores[i] for i in range(len(test_scores)-1)]

# Create line chart
plt.figure(figsize=(8, 4))
plt.plot(months, test_scores, marker='o', color='b', linestyle='-')

# Add labels and title
plt.xlabel('Month', fontsize=14)
plt.ylabel('Test Score', fontsize=14)
plt.title('UNT test Scores Over Time', fontsize=16)

# Hide y-axis
plt.gca().axes.get_yaxis().set_visible(False)

# Add text annotations for score changes
for i, txt in enumerate(score_changes):
    plt.annotate(f'+{txt}', (months[i+1], test_scores[i+1]), textcoords="offset points", xytext=(0,10), ha='center')

# Show plot
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
from matplotlib.collections import LineCollection
import numpy as np

# Data
months = ['October', 'January', 'March', 'May', 'June']
test_scores = [69, 101, 116, 128, 129]

# Define start and end colors
start_color = np.array([8, 27, 140]) / 255.0  # RGB values for #081b8c
end_color = np.array([130, 11, 92]) / 255.0  # RGB values for #820b5c

# Interpolate colors along the gradient
colors = [tuple((start_color * (1 - t) + end_color * t).tolist()) for t in np.linspace(0, 1, len(test_scores)-1)]

# Create line segments with varying colors
segments = [((months[i], test_scores[i]), (months[i+1], test_scores[i+1])) for i in range(len(test_scores)-1)]

# Plot the line segments with gradient colors
plt.figure(figsize=(8, 5))
for i in range(len(segments)):
    x_values, y_values = zip(*segments[i])
    plt.plot(x_values, y_values, color=colors[i], linewidth=2)

# Add text annotations for score changes
for i in range(len(months)-1):
    plt.text(months[i+1], test_scores[i+1], f'+{score_changes[i]}', ha='center', va='bottom')

# Add labels and title
plt.xlabel('Month', fontsize=14)
plt.ylabel('Test Score', fontsize=14)
plt.title('Test Scores Over Time', fontsize=16)




# Show plot
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
months = ['October', 'January', 'March', 'May', 'June']
test_scores = [69, 101, 116, 128, 129]

# Calculate differences between consecutive test scores
score_changes = [test_scores[i+1] - test_scores[i] for i in range(len(test_scores)-1)]

# Create a gradient color
colors = plt.cm.coolwarm(np.linspace(0, 1, len(months)))

# Create line chart
plt.figure(figsize=(10, 6))
for i in range(len(months)-1):
    plt.plot(months[i:i+2], test_scores[i:i+2], marker='o', color=colors[i], linestyle='-')

# Add text annotations for score changes
for i in range(len(months)-1):
    plt.text(months[i+1], test_scores[i+1], f'+{score_changes[i]}', ha='center', va='bottom')

# Add labels and title
plt.xlabel('Month', fontsize=14)
plt.ylabel('Test Score', fontsize=14)
plt.title('Test Scores Over Time', fontsize=16)



# Show plot
plt.grid(True)
plt.tight_layout()
plt.show()
```
